## **Plot in multiple axes**

**`fig.subplots_adjust(hspace=)`**: set the hspace between two axes

**`set_xlabel(fontsize= )`**: set the fontsize of  x label

**`set_ylabel(fontsize= )`**: set the fontsize of  y label

**`set_title(fontsize= )`**: set the fontsize of the title

**`ax.set_facecolor('#aa0099'`**): set the background color of an axes

**`fig, ax = plt.subplots(rowNum, colNum, figsize=(w, h))`**


```python
# plot multiple axes
days = list(range(1, 9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]

# create a figure that contains two vertical axes
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

# set the virtical space between the axes
fig.subplots_adjust(hspace=0.4)

# set the background color
#ax[0].set_facecolor('gray')
ax[0].set_facecolor('#ffffee')
ax[1].set_facecolor('#fefefe')

#plot celsisu_min and celsius_max in two separate plots
ax[0].plot(days, celsius_min)
ax[1].plot(days, celsius_max)

# set titles and fontsize.
ax[0].set_title('Celsius Min', fontsize=10)
ax[1].set_title('Celsium Max')

#set labels and font size
ax[0].set_xlabel('Day', fontsize=10)
ax[1].set_xlabel('Day', fontsize=20)

#set legend
ax[0].legend(['Min Temp'])
ax[1].legend(['Max Temp'])

plt.show()
```

# **Bar Chart**

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.
*  The bar plots can be plotted horizontally or vertically
*  A bar chart describes the comparisons between the discrete categories. It can be created using the bar() method.

Syntax:

***plt.bar(x, height, width, bottom, align, color)***
* x:  the x-coordinates of the bars. It can be a single value (for a single bar) or a sequence of values (for multiple bars).
* height: the height(s) of the bars. It can be a single value or a sequence, corresponding to the x values.
* width: the width(s) of the bars. The default value is 0.8. It can be a single value or a sequence.
* bottom: the y-coordinate(s) of the base of the bars. The default value is 0. This is useful for creating stacked bar charts. It can be a single value or a sequence.
* align: controls the alignment of the bars relative to their x coordinates. It can be either:
 * `center` (default): Centers the base of the bar on the x position.
 * `edge`: Aligns the left edge of the bar with the x position.
* color: specifies the color(s) of the bars. It can be a single color name or hexadecimal code (e.g., 'red', '#FF0000'), or a sequence of colors for multiple bars.



```python
# data to display on plots
x = [3, 1, 5, 12, 2, 9, 4]
y = [3, 2, 1, 4, 5, 6, 7]

# This will plot a simple bar chart
plt.bar(x, y)
#plt.bar(x, y, color=['red','yellow','gray','blue','green','purple'])
#plt.bar(x, y, bottom=3)

#plt.bar(x, y, align='edge')
# Title to the plot
plt.title("Bar Chart")

# Adding the legends
plt.legend(["bar"])
plt.show()
```

# **Histogram**

A Histogram represents the distribution of a numeric variable for one or several groups. The values are split in **bins**, each bin is represented as a bar.


Bins are clearly identified as consecutive, non-overlapping intervals of variables. The hist() function is used to compute and create histogram of x.

**Bins:**
*   If an integer is given, bins + 1 bin edges are calculated and returned.
*   If bins is a sequence, gives bin edges, including left edge of first bin and right edge of last bin. In this case, bins is returned unmodified.

```python
import matplotlib.pyplot as plt
# data to display on plots
x = [1, 2, 3, 4, 5, 6, 7, 4, 4, 5]
# This will plot a simple histogram
fig, ax=plt.subplots()
ax.hist(x, bins = [1, 2, 3, 4, 5, 6, 7, 8], rwidth=0.6)
#ax.hist(x, bins = 4, rwidth=0.5)
# Title to the plot
ax.set_title("Histogram")
# Adding the legends
ax.legend(["bar"])
plt.show()
```

# **Pie Chart**

A Pie Chart is a circular statistical plot that can display only one series of data.

The area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data.

The slices of pie are called **wedges**.

```python
from matplotlib import pyplot as plt
import numpy as np


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)

# show plot
plt.show()
```

#### **Add labels to Chart**

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]


# Creating explode data
explode = (0.1, 0.0, 0.2, 0.0, 0.0, 0.0)

# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = cars,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))

# Adding legend
ax.legend(wedges, cars,
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight ="bold")
plt.setp(texts, size = 12, weight ="bold")
ax.set_title("Customizing pie chart")

# show plot
plt.show()
```
