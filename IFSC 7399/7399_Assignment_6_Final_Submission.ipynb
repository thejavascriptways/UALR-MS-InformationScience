**Before starting, run the code below to import all necessary functions and libraries.**


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Run the code to create DataFrame df.
url = "https://raw.githubusercontent.com/the-codingschool/TRAIN/main/tokyo_olympics/tokyo_2021_cleaned.xlsx"
df = pd.read_excel(url)
df.head()
```

**Problem 1**

Create a scatter plot with `Gold Medal` on the x-axis and `Bronze Medal` on the y-axis. Make sure to include title, xlabel, and ylabel.

```python
# specify the data frame to be used along with x and y axis column names
plt.scatter(data=df, x='Gold Medal', y='Bronze Medal')

# decorate the graph
plt.xlabel('Gold Medal')
plt.ylabel('Bronze Medal')
plt.title('Assignment 6 Scatter Plot')

# display the scatter plot
plt.show()
```

**Problem 2**

Create a line plot with `Rank` on the x-axis and `Total` on the y-axis. Remember to title and label your plot.

```python
# specify the data frame to be used along with x and y axis column names
df.plot(x='Rank', y='Total', kind='line')

# decorate the graph
plt.xlabel('Rank')
plt.ylabel('Total')
plt.title('Assignment 6 Line Plot')

# display the line plot
plt.show()
```

**Problem 3**

Create a bar plot with `Continent` on the x-axis and `Total` on the y-axis. Remember to title and label your plot.

```python
# specify the data frame to be used along with x and y axis column names
df.plot(x='Continent', y='Total', kind='bar')

# decorate the graph
plt.xlabel('Continent')
plt.ylabel('Total')
plt.title('Assignment 6 Bar Plot')

# display the bar plot
plt.show()
```
