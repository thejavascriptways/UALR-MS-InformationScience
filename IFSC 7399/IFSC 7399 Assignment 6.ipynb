**Before starting, run the code below to import all necessary functions and libraries.**


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Run the code to create DataFrame df.
url = "https://raw.githubusercontent.com/the-codingschool/TRAIN/main/tokyo_olympics/tokyo_2021_cleaned.xlsx"
df = pd.read_excel(url)
df.head()
```

**Problem 1**

Create a scatter plot with `Gold Medal` on the x-axis and `Bronze Medal` on the y-axis. Make sure to include title, xlabel, and ylabel.

```python
#plot type and data
plt.scatter(df['Gold Medal'], df['Bronze Medal'])

#plot title
plt.title('Gold Medal vs Bronze Medal')

#plot x and y labels
plt.xlabel('Gold Medal')
plt.ylabel('Bronze Medal')

#display
print("Scatter plot of Gold medal vs Bronze Medal \n")
plt.show()
```

**Problem 2**

Create a line plot with `Rank` on the x-axis and `Total` on the y-axis. Remember to title and label your plot.

```python
#plot type and data

plt.plot(df['Rank'], df['Total'])

#plot title
plt.title('Rank vs Total')

#plot x and y labels
plt.xlabel('Rank')
plt.ylabel('Total')

print("Line plot of Rank vs Total \n")
plt.show()

```

**Problem 3**

Create a bar plot with `Continent` on the x-axis and `Total` on the y-axis. Remember to title and label your plot.

```python
# plot type and data

plt.bar(df['Continent'], df['Total'])

#plot title
plt.title('Continent vs Total')

#plot x and y labels
plt.xlabel('Continent')
plt.ylabel('Total')

print("Bar plot of Continent vs Total \n")
plt.show()
```
